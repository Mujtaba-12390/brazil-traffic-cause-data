# Taking Correalation 

```python
df=traffic.corr()["Slowness in traffic (%)"].sort_values(ascending=False)
```

```python
df=df.drop("Hour (Coded)")
```

```python
df
```

# `Result `
    According to the given data Lack of electricity is the cause of traffic jam and Broken Truck are the very week correlate with slowness and the total truck are increase in number but in reality Lack of electricity is the cause of Traffic jam.The below heatmap also showed this.

```python
slowness_correlation_df = df.to_frame()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(slowness_correlation_df, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Cause of Slowness in Traffic (%)")
plt.savefig("Final Result.png")
plt.show()
```
